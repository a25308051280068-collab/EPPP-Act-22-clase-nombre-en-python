#### Nombre de clase conductor

```python
class Conductor:
    # El método __init__ define los atributos (datos)
    def __init__(self, nombre, placas, edad):
        self.nombre = nombre
        self.placas = placas
        self.edad = edad

    # Definición de las Funciones (Métodos)
    def transportar(self):
        print(f"El conductor {self.nombre} está transportando pasajeros en el vehículo {self.placas}.")

    def planificacion_de_rutas(self):
        print(f"Calculando la mejor ruta para evitar el tráfico... (Edad del conductor: {self.edad} años).")

    def mantenimiento_de_vehiculo(self):
        print(f"El vehículo con placas {self.placas} ha entrado a revisión técnica.")

# --- Lógica con Input para crear los objetos ---

print("--- REGISTRO DE CONDUCTORES ---")

# Datos para el Conductor 1
print("\nIngrese los datos para el Conductor 1:")
nombre1 = input("Nombre: ")
placas1 = input("Placas del vehículo: ")
edad1 = input("Edad: ")
conductor1 = Conductor(nombre1, placas1, edad1)

# Datos para el Conductor 2
print("\nIngrese los datos para el Conductor 2:")
nombre2 = input("Nombre: ")
placas2 = input("Placas del vehículo: ")
edad2 = input("Edad: ")
conductor2 = Conductor(nombre2, placas2, edad2)

# --- Ejecución de las funciones ---

print("\n" + "="*30)
print(f"--- Actividades de {conductor1.nombre} ---")
conductor1.transportar()
conductor1.planificacion_de_rutas()
conductor1.mantenimiento_de_vehiculo()

print("\n" + "="*30)
print(f"--- Actividades de {conductor2.nombre} ---")
conductor2.mantenimiento_de_vehiculo()
conductor2.transportar()
conductor2.planificacion_de_rutas()

print("\n" + "="*30)
print("Programa terminado")
print("Programa realizado por Christian Garcia NL = 21")
```
